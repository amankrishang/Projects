# 1.Survived vs Sex

```python
grouped_data = df.groupby(["Sex", "Survived"]).size().unstack()
grouped_data
```

```python
grouped_data.plot(kind= "bar")
plt.title("Sex vs Survived")
plt.ylabel("Number of people")
plt.show()
```

# survived vs age

```python
df[df["Survived"]==1]["Age"]
```

```python
sns.kdeplot(df[df["Survived"]==1]["Age"], color = "green", label = "Survived")
sns.kdeplot(df[df["Survived"]==0]["Age"], color = "red", label = "Not Survived")
plt.legend()
plt.show()
```

```python
grouped_data = df.groupby(["Embarked", "Survived"]).size().unstack()
grouped_data
```

```python
grouped_data.plot(kind= "bar")
plt.title("Embarked vs Survived")
plt.ylabel("Number of people")
plt.show()
```

```python
sns.kdeplot(df[df["Survived"]==1]["Fare"], color = "green", label = "Survived")
sns.kdeplot(df[df["Survived"]==0]["Fare"], color = "red", label = "Not Survived")
plt.legend()
plt.show()
```

# age vs sex

```python
male_data = df[df["Sex"]=="male"]
female_data = df[df["Sex"]=="female"]
male_data.head(3)
```

```python
female_data.head(3)
```

```python
sns.histplot(data = male_data, x="Age", color='blue', label = "Male", alpha = 1)
sns.histplot(data = female_data, x="Age", color='pink', label = "Female", alpha = 1)
plt.legend()
plt.show()
```
